# Making this Project using `pytesseract`
Making a OCR powered reading system, that will read aloud the text in the image.
steps:
- Use `pytesseract` to extract text from the image 
- use `pyttsx3` to read aloud the text from the image

```python
# standard imports
import cv2
import pytesseract
import pyttsx3
```

### Setting up the tesseract command line

```python
pytesseract.pytesseract.tesseract_cmd = 'C:\\Program Files\\Tesseract-OCR\\tesseract.exe'
```

### Initializing the `pyttsx3` speaking engine

```python
speaker = pyttsx3.init()
```

### defining a function which will predict the string in the picture and speak aloud the sentence

```python
def predict_and_speak(image) -> None:
    '''Takes images as input and says what is written in it.'''
    text = pytesseract.image_to_string(image)
    
    speaker.say(text)
    speaker.runAndWait()
    return text
```

### Testing 

```python
import matplotlib.pyplot as plt
```

```python
def plot_image(image, title:str='image'):
    '''Plots the image'''
    plt.imshow(image)
    plt.axis('off')
    plt.title(title)
```

```python
image = cv2.imread('./images/sample3.jfif')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plot_image(image)
```

```python
print(u'\033[31m' + predict_and_speak(image))
```
